```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Construct Plot
N = 100
l = 0.5
amp = 1.
t = np.linspace(0,100,N)
A = np.concatenate((amp*np.sin(t[:N/2]*l)+2,amp*np.sin(t[N/2:]*l)+8))
B = np.concatenate((amp*np.sin(t[:N/2]*l)+8,amp*np.sin(t[N/2:]*l)+2))
plt.plot(t,A,t,B)
plt.title('Correlation Example')
plt.legend(['A','B'],loc=0)
plt.ylim(0,10)
plt.show()


# Calculate Coefficients
stdA = np.std(A)
stdB = np.std(B)
cov = np.cov(A,B)
corr = np.corrcoef(A,B)
print 'std(A) = {0:.3f}, std(B) = {1:.3f}\ncovarience = {2:.3f}\ncorrelation = {3:.3f}'.format(stdA,stdB,cov[0,1],corr[0,1])

```
